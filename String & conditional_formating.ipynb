# String

```python
#string ---> string is a data type that stores a sequence of characters.

# Whatever we write inside  "" these inverted comma called string.

# WE can write strings in three ways ("",'',""" """).

# Strings are immutable we cannot add & remove elements in orginal string.
# When we use string methods the result shows us in new string.
# Change has not made in original string.
```

```python

# concatenation
a = "Mayank"
b = "singh"
c = a + b
print(c)


#length of string -->  it counts the length


print(len(c)) 

```

```python
#Indexing: indexing is like a position numbers
#it always starts with 0.


A = "Mayank"

print(A[0])
print(A[1])
print(A[5])


# We can only access the string with the help of indexing.
# We cannot replace or assign the new values .


# Slicing --> Accessing part of a string to get the substring.
# starding index included.
# ending index not icluded.


A = "Rajeshwari"

print(A[0:6])
print(A[6:])
print(A[:4])
print(A[::2]) #starting-ending-step
print(A[::-1]) #Negative 



#Negative indexing

A = "Rajeshwari"

print(A[-5])
print(A[-9])


# Negative slicing

A = "Rajeshwari"

print(A[-4:])


```

```python
#Methods of string

A = "rajeshwari"

print(A.endswith("i"))

print(A.capitalize())

print(A.replace("a","o"))

print(A.find("a")) #it tell us idx no of first occurance

print(A.find("m")) #it return the -1

print(A.count("a")) #it tell us the counts of word

```

# Conditional_Formatting

```python
#(if--elif---else)

# conditional statement ---> are used in programming to make precise decisions.

#ex:

age = 2

if(age >= 18):
    print("can vote")
else:
    print("cannot_vote")
```

```python
#if--elif--else
# Evertime before using elif there must an if.
# We can use multiple if,elif together.


light = "Green"

if(light == "red"):
    print("Stop")
elif(light == "Green"):
    print("Go")
elif(light == "yellow"):
    print("Wait")
else:
    print("Light is broken")
```

```python
light = "red"
if(light == "red"):  #if is getting true here
    print("stop")
if(light == "red"):
    print("go")
if(light == "red"):
    print("rajeshwari")

#in this condition if will check every condition after being true also.
# so if is printing all coditions values.
```

```python
#nested if

age = 90

if(age >= 18):
    if(age >= 60): 
        print("cannot drive")
    else:
        print("Can drive") 
        
```
