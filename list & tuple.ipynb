# List & Tuple

```python
# list = A built in data type that stores sequence of values.

# it can store elements of different type (int,float,string etc..)




list = [99,79.9,0,True]
print(type(list))
```

```python
#indexing in list
#indexing always starts from 0.


list = [99,88,77,66,55,44,33,22,11]

print(list[5]) #we have to fetch 44
print(list[8])

#length of list

print(len(list)) #len always starts from 1
```

```python
#slicing : to get the sublist.
# slicing ending index doesn't include.
list = [99,88,77,66,55,44,33,22,11]


# we have to fetch 66 : 22

print(list[3:8])

# 66 to 11

print(list[3:])

# 99 to 66

print(list[:4])

# 99 to 11
print(list[:])
#stepping
print(list[::2])

#reversing the list 

print(list[::-1])
```

```python
#Negative indexing & slicing
# negative indexing always starts from -1.
list = [99,88,77,66,55,44,33,22,11]


print(list[-6])


#66 to 22 from negative slicing

print(list[-6:-1])

# 66 to 11

print(list[-6:])

#99 to 66

print(list[-9 : - 5])


#reverse
print(list[::-1])


```

```python

# list is Mutable 
# in list we can access the element & we can also change the element


student = ["karan",95.4,19,"Delhi"]

student[0]
student[1] = 99 #we are accessing the element & assigning new value

```

```python
# methods of list : list specific methods

list = [40,45,33,22,11]

list.append(400) # adding at the end

list.sort() # it will sort list in ascending

list.sort(reverse = True) #it will sort list in descending

list.reverse() # it will reverse the list

list.insert(2,500) #insert the new value at given place

list.remove(500) #first occurence value removed only

list.pop() #it removes last element of list

```

# Tuple

```python
# tuple : it creates immutable sequence of values.

Tup = (2,1,3,4,5)

# tuples are immutable:

# Tup[0] = 200 this is not possible in tuble
# we can not assign values to tuple
# We can only access the element

Tup[0]
Tup[1]
Tup[2]

```

```python
# Empty tuple

tup = () #this is an empty tuple

print(type(tup))


#single value tuple

tup = (99,) 
print(type(tup))
```

```python
#tuples method:

tup = ("mayank",100,44.5,True,True)

print(tup.index(True)) # it will only fetch first accurance index.

print(tup.count(True)) #it will count all occurance of an element.

```

```python
#Palindrome.

# palindrome are those things which are same if we read from start or end.

# ex: (2,1,2) (ma'am) (racecar)


# copy (2,1,2)
# reverse = if same as it is.

a = [2,1,2,3]
b = a.copy()   
b.reverse()

if(a == b):
    print("palindrome")
else:
    print("not palindrome")


    
# word = "madam"

# if(word == word[::-1]):
#     print("palindrome")
# else:
#     print("not palindrome")


```
